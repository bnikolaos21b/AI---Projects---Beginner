We start by importing essential libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

### Loading the Dataset
The FER (Facial Expression Recognition) dataset contains 48x48 grayscale images of human faces, each associated with an emotion label (e.g., happy, sad, angry). We load this dataset from a CSV file where each image is represented as a series of pixel values, and the emotion labels are numeric. This data serves as the foundation for training our CNN model to classify facial expressions.


```python
# Load the dataset
data = pd.read_csv("/home/nikos_pc/AI_Beginner_Projects/Face Expression Recognition/data/fer2013/fer2013/fer2013.csv")
data.head()


num_classes = data['emotion'].nunique()
print(f"Number of emotion classes: {num_classes}")


emotion_labels = {
    0: 'Angry',
    1: 'Disgust',
    2: 'Fear',
    3: 'Happy',
    4: 'Sad',
    5: 'Surprise',
    6: 'Neutral'
}
```

### Preprocessing the Images
Each image is stored as a string of pixel values. We must convert these strings into arrays of floats and reshape them into the original 48x48 pixel grid. This step is crucial because neural networks expect input data in a specific numerical format, and reshaping prepares the images for convolutional processing.


```python
def preprocess_pixels(pixels):
    pixels = np.array(pixels.split(), dtype='float32')
    return pixels.reshape(48, 48)

data['pixels'] = data['pixels'].apply(preprocess_pixels)

X = np.stack(data['pixels'].values)
y = data['emotion'].values
```

### Normalizing the Pixel Data
Pixel values in the images range from 0 to 255. To make the model training more efficient and to prevent large gradients, we normalize the pixel values by dividing by 255.0. This scales the pixel values between 0 and 1, which helps the network learn faster and reduces the risk of gradient-related issues during optimization.


```python
X = X / 255.0

X = X.reshape(-1, 48, 48, 1)

y = to_categorical(y, num_classes=num_classes)

train_idx = data['Usage'] == 'Training'
val_idx = data['Usage'] == 'PublicTest'
test_idx = data['Usage'] == 'PrivateTest'

X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val = X[val_idx], y[val_idx]
X_test, y_test = X[test_idx], y[test_idx]

print(f"Training samples: {X_train.shape[0]}")
print(f"Validation samples: {X_val.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
```

### Data Augmentation
Data augmentation generates new, slightly modified versions of images to improve the model’s ability to generalize. By rotating, zooming, and flipping images, we expose the model to varied representations of the same data, helping it become more robust and less prone to overfitting. This is especially important when working with a relatively small dataset, as augmentation effectively increases the diversity of the training data.


```python
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)

datagen.fit(X_train)
```

### Building the CNN Model
We construct our Convolutional Neural Network (CNN) to detect patterns in the input images:
- **Convolutional layers**: These layers apply filters to the input images to detect features like edges and textures.
- **MaxPooling layers**: These reduce the spatial dimensions of the feature maps, making the model more computationally efficient while retaining essential features.
- **Dense layers**: After feature extraction, fully connected layers learn to combine these features to classify the image into one of the emotion classes.
By stacking convolutional and pooling layers, we enable the model to learn increasingly complex patterns from the images.


```python
model = Sequential()

# First convolutional block
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(48,48,1)))
model.add(MaxPooling2D(2,2))

# Second convolutional block
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

# Third convolutional block
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

# Flatten and Dense layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Dropout for regularization
model.add(Dense(num_classes, activation='softmax'))

model.summary()
```

### Compiling the Model
To train the model, we use the Adam optimizer, which adapts the learning rate during training for better performance. We also use **categorical cross-entropy** as our loss function, which is ideal for multi-class classification problems like this one. By monitoring accuracy as a metric, we can track how well the model is performing during training and validation.


```python
model.compile(optimizer=Adam(learning_rate=0.0001), 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

### Training the Model
We now train our model using the augmented data. The model will iterate over the dataset for 100 epochs, learning to minimize the error in predicting facial expressions. We also validate the model’s performance on a separate validation set after each epoch. Monitoring both training and validation performance helps us detect if the model is overfitting or underfitting.


```python
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=64),
    epochs=100,
    validation_data=(X_val, y_val)
)

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy*100:.2f}%")

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

## Evaluation and Conclusions

The training process for this facial expression recognition model demonstrated a steady improvement in accuracy and reduction in loss over the course of 100 epochs. The initial accuracy of the model started at 24.21%, with a corresponding loss of 1.8323. Over time, both training and validation accuracy increased, showing that the model was learning effectively.

The validation accuracy improved from 28.70% at epoch 1 to 58.37% by epoch 100, with a validation loss decreasing from 1.7503 to 1.1305. The final test accuracy achieved was 60.07%, indicating that the model can generalize reasonably well to unseen data.

Despite these improvements, the overall accuracy suggests that there is still room for enhancement. Potential next steps could include:

- **Data augmentation**: Adding transformations to the training images to improve generalization.
- **Model architecture improvements**: Adding more layers or tweaking existing ones to capture more complex patterns.
- **Additional training**: Extending the number of epochs or implementing early stopping based on validation loss.

Overall, this model lays a solid foundation for recognizing facial expressions, but further refinement and optimization could lead to even better performance.
